```python
#I will choose to drop all rows containing missing values since it only amounts for a 5.8%, with the exception of vehicle_type
#which I will process differently
filtered=file.dropna(subset=['state', 'response', "months_since_last_claim", "number_of_open_complaints", "vehicle_class", "vehicle_size"])
```

```python
#VEHICLE TYPE
file.vehicle_type.value_counts(dropna=False)
file["vehicle_type"].unique()
```

```python
#There is just one correctly defined vehicle type, A. We will assign the remaining to vehicle type B, we would lose too many
#observations if we dropped all those rows.
filtered['vehicle_type'] = filtered['vehicle_type'].fillna('B').copy()
filtered.vehicle_type.value_counts(dropna=False)
```

```python
#Checking for remaining missing values
nulls_df = pd.DataFrame(round(filtered.isna().sum()/len(filtered),4)*100)
#display(nulls_df)
nulls_df = nulls_df.reset_index()

nulls_df.columns = ['header_name', 'percent_nulls']
display(nulls_df)
nulls_df.loc[(nulls_df['percent_nulls'] >0)]
```

```python
# 6. Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only 
#the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, 
#consider only January and February.

#Firstly we adapt the data from effective_to_date into datetime format  
filtered['effective_to_date'] = pd.to_datetime(filtered['effective_to_date'], errors='coerce') 
filtered.dtypes
display(filtered.head())
```

```python
# Show a plot of the total number of responses
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.catplot(data=filtered, x="response", kind="count")
```

```python
#Show a plot of the response rate by the sales channel
#Firstly, I createa proportion tale for the different sales channels
cross_tab_prop = pd.crosstab(index=filtered['sales_channel'],
                             columns=filtered['response'],
                             normalize="index")
display(cross_tab_prop)

cross_tab_prop.plot(kind='bar', 
                    stacked=True, 
                    colormap='tab10', 
                    figsize=(10, 6))

plt.legend(loc="upper left")
plt.xlabel("Sales Channel")
plt.ylabel("Proportion")
plt.show()

```
